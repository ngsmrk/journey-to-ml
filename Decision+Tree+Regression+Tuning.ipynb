# Example of tuning decision tree for regression

Changes `max_depth` and `min_samples_leaf`.

## Load data

```python
from pandas import read_csv
path = "https://oml-data.s3.amazonaws.com/" # load data from http location; you can also load from local path
data = read_csv(path + "boston-housing.csv", index_col=0)
```

Inspect data with `pandas_profiling`

```python
from pandas_profiling import ProfileReport
pfr = ProfileReport(data)
pfr
```

Prepare `X` and `y` for scikit

```python
target_column = 'medv'
outputs = data[target_column]
y = outputs.values

features = data.drop(target_column, axis=1)
X = features.values
```

## Random extract of train and test sets

```python
TEST_SIZE = 0.3 # ratio of data to have in test
SEED = 8 # to be used to initialize random number generator, for reproducibility
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=TEST_SIZE)
                                                    #, random_state=SEED)
```

Quick check on what we just got...

```python
print(X_train.shape)
print(X_test.shape)
print(X.shape)
```

## Decision Tree Example

Tries to find optimal `max_depth` and `min_samples_leaf` for DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
import sklearn.metrics as metrics
results = []
for depth in range(1,10):
    for min_samples in range(1,10):
        # model = DecisionTreeRegressor(random_state=SEED, max_depth=x)
        model = DecisionTreeRegressor(max_depth=depth, min_samples_leaf=min_samples)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        result = (depth, min_samples, metrics.r2_score(y_test, y_pred))
        results.append(result)
```

```python
def max(results):
    max = -1    
    winner = None
    for tuple in results:
        print(tuple)
        r2_score = tuple[2]
        if (r2_score > max):
            max = r2_score
            winner = tuple
        
    print("Winner " + str(winner))
    return winner
```

```python
max(results)
```

## Random Forest Example

Tries to find optimal `max_depth` and `min_samples_leaf` for RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
import sklearn.metrics as metrics
rf_results = []
for depth in range(1,10):
    for min_samples in range(1,10):
        # model = DecisionTreeRegressor(random_state=SEED, max_depth=x)
        model = RandomForestRegressor(n_estimators=10, max_depth=depth, min_samples_leaf=min_samples)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        result = (depth, min_samples, metrics.r2_score(y_test, y_pred))
        rf_results.append(result)
        
max(rf_results)        
```

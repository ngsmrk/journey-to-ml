## Load data

```python
import pandas as pd
data = pd.read_csv("/data/kaggle-give-me-credit-nomissing.csv", index_col=0)
```

Inspect data with `pandas_profiling`

```python
from pandas_profiling import ProfileReport
pfr = ProfileReport(data)
pfr
```

Prepare `X` and `y` for scikit

```python
target_column = 'SeriousDlqin2yrs'
outputs = data[target_column]
y = outputs.values
print(y)
```

```python
features = data.drop(target_column, axis=1)
# features = features.drop("id", axis=1)
X = features.values
print(X)
```

## Random extract of train and test sets

```python
TEST_SIZE = 0.3 # ratio of data to have in test
SEED = 8 # to be used to initialize random number generator, for reproducibility
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=TEST_SIZE,
                                                    random_state=SEED)
```

Quick check on what we just got...

```python
print(X_train.shape)
print(X_test.shape)
print(X.shape)
```

## Initialize estimator

Run one of these 2 cells below:

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=SEED)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=SEED)
```

## Fit model on train and predict on test

```python
%time model.fit(X_train, y_train)
%time y_pred = model.predict(X_test)
```

(`%time` is an IPython 'magic' which can be added at the beginning of a line of code)

```python
sum(abs(y_pred - y_test)) / len(y_pred)
```

```python
def mae(outputs, predictions):
    diff = sum(abs(predictions - outputs))
    return diff / len(outputs)

mae(y_test, y_pred)
```

```python
import sklearn.metrics as metrics
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
metrics.precision_score(y_test, y_pred)
```
